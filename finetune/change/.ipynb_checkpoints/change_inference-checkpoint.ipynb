### Model Loading

```python
def get_model(chesapeake_checkpoint_path, clay_checkpoint_path, metadata_path):
    model = ChesapeakeSegmentor.load_from_checkpoint(
        checkpoint_path=chesapeake_checkpoint_path,
        metadata_path=metadata_path,
        ckpt_path=clay_checkpoint_path,
    )
    model.eval()
    return model

```

### Data Preparation

```python
def get_data(
    train_before_chip_dir,
    train_after_chip_dir,
    train_label_dir,
    val_before_chip_dir,
    val_after_chip_dir,
    val_label_dir,
    metadata_path,
    batch_size,
    num_workers,
    platform,
):
    # Initialize ChesapeakeDataModule with both before and after chip directories
    dm = ChesapeakeDataModule(
        train_before_chip_dir=train_before_chip_dir,
        train_after_chip_dir=train_after_chip_dir,
        train_label_dir=train_label_dir,
        val_before_chip_dir=val_before_chip_dir,
        val_after_chip_dir=val_after_chip_dir,
        val_label_dir=val_label_dir,
        metadata_path=metadata_path,
        batch_size=batch_size,
        num_workers=num_workers,
        platform=platform,
    )
    
    # Setup dataloaders for the 'fit' stage
    dm.setup(stage="fit")
    
    # Fetch a batch from the validation dataloader
    val_dl = iter(dm.train_dataloader())
    batch = next(val_dl)
    
    # Get metadata
    metadata = dm.metadata
    
    # Return both the batch and metadata
    return batch, metadata
```

### Prediction

```python
def run_prediction(model, before_datacube, after_datacube):
    with torch.no_grad():
        # Pass both before and after datacubes to the model
        outputs = model(before_datacube=before_datacube, after_datacube=after_datacube)
        print(f"Output shape: {outputs.shape}")
    
    # Optionally, resize the outputs
    outputs = F.interpolate(
        outputs, size=(224, 224), mode="bilinear", align_corners=False
    )
    
    return outputs
```

### Post-Processing

```python
def denormalize_images(normalized_images, means, stds):
    means = np.array(means).reshape(1, -1, 1, 1)
    stds = np.array(stds).reshape(1, -1, 1, 1)
    denormalized_images = normalized_images * stds + means
    denormalized_images = (denormalized_images / 3000) * 255
    return denormalized_images.astype(np.uint8)  # Do for NAIP/LINZ


def post_process(batch, outputs, metadata):
    # Extract the first band (class) from the model output
    first_band = outputs[:, 0, :, :]
    
    # Apply sigmoid function to the first band
    sigmoid_band = torch.sigmoid(first_band)
    
    # Convert to numpy array
    preds = sigmoid_band.detach().cpu().numpy()
    
    # Extract labels and input images from the batch
    labels = batch["label"].detach().cpu().numpy()[:, 0, :, :]
    
    # Access "pixels" in the before and after datacubes
    before_pixels = batch["before"]["pixels"].detach().cpu().numpy()
    after_pixels = batch["after"]["pixels"].detach().cpu().numpy()

    # Denormalize both "before" and "after" images
    means = list(metadata["sentinel-2-l2a"].bands.mean.values())
    stds = list(metadata["sentinel-2-l2a"].bands.std.values())
    
    before_norm_pixels = denormalize_images(before_pixels, means, stds)
    after_norm_pixels = denormalize_images(after_pixels, means, stds)

    # Extract RGB channels for before and after images
    before_images = rearrange(before_norm_pixels[:, :3, :, :], "b c h w -> b h w c")
    after_images = rearrange(after_norm_pixels[:, :3, :, :], "b c h w -> b h w c")

    return before_images, after_images, labels, preds


```

### Plotting

```python
def plot_predictions(before_images, after_images, labels, preds):
    cmap = plt.cm.viridis
    num_images = len(before_images)
    
    fig, axes = plt.subplots(4, num_images, figsize=(20, 16))
    
    for i in range(num_images):
        # Plot the "before" image
        axes[0, i].imshow(before_images[i])
        axes[0, i].axis("off")
        axes[0, 0].set_title("Before", fontsize=12)
        
        # Plot the "after" image
        axes[1, i].imshow(after_images[i])
        axes[1, i].axis("off")
        axes[1, 0].set_title("After", fontsize=12)
        
        # Plot the label image (ground truth)
        im = axes[2, i].imshow(labels[i], cmap=cmap, vmin=0, vmax=1)
        axes[2, i].axis("off")
        axes[2, 0].set_title("Label", fontsize=12)
        plt.colorbar(im, ax=axes[2, i], fraction=0.046, pad=0.04)

        # Plot the prediction
        im = axes[3, i].imshow(preds[i], cmap=cmap, vmin=0, vmax=1)
        axes[3, i].axis("off")
        axes[3, 0].set_title("Prediction", fontsize=12)
        plt.colorbar(im, ax=axes[3, i], fraction=0.046, pad=0.04)
    
    plt.tight_layout()
    plt.show()



def plot_data(ax, data, row_offset, title=None, cmap=None):
    for i, item in enumerate(data):
        row = row_offset + (i // 8)
        col = i % 8
        im = ax[row, col].imshow(item, cmap=cmap, vmin=0, vmax=1)
        ax[row, col].axis("off")
        if title and col == 0:
            ax[row, col].set_title(title, rotation=0, fontsize=12)
        if cmap is not None:
            plt.colorbar(im, ax=ax[row, col], fraction=0.046, pad=0.04)
```

```python
# Load model
model = get_model(CHESAPEAKE_CHECKPOINT_PATH, CLAY_CHECKPOINT_PATH, METADATA_PATH)
```

```python
# Get data with before and after chip directories
batch, metadata = get_data(
    train_before_chip_dir=train_before_chip_dir,
    train_after_chip_dir=train_after_chip_dir,
    train_label_dir=train_label_dir,
    val_before_chip_dir=val_before_chip_dir,
    val_after_chip_dir=val_after_chip_dir,
    val_label_dir=val_label_dir,
    metadata_path=METADATA_PATH,
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKERS,
    platform=PLATFORM,
)

# Move the batch to GPU (for both before and after images)
batch['before'] = {k: v.to("cpu") for k, v in batch['before'].items()}
batch['after'] = {k: v.to("cpu") for k, v in batch['after'].items()}
batch['label'] = batch['label'].to("cpu")  # Assuming the label is also part of the batch
```

```python
# Run prediction with both before and after datacubes
outputs = run_prediction(model, batch['before'], batch['after'])
print(outputs.shape)
# Post-process the results
before_images, after_images, labels, preds = post_process(batch, outputs, metadata)
```
